```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv("shopping_trends.csv")
df = pd.DataFrame(data)
print(df)
print(df.describe())

df.rename(columns={'Purchase Amount (USD)':'Current Purchases'}, inplace=True)
long_data = pd.melt(df, id_vars=['Season'], value_vars=['Current Purchases', 'Previous Purchases'],var_name='Purchase Type', value_name='Amount (USD)')


plt.figure(figsize=(8, 6))
sns.barplot(x='Season', y='Amount (USD)', hue='Purchase Type', data=long_data)


plt.title('Comparison of Purchase Amounts by Season')
plt.xlabel('Season')
plt.ylabel('Amount (USD)')

plt.show()
```

```python
gender_totals = df.groupby('Gender')['Current Purchases'].sum()
plt.figure(figsize=(8, 6))
plt.pie(gender_totals, labels=gender_totals.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Current Purchases by Gender')
plt.show()
gender_totals = df.groupby('Gender')['Previous Purchases'].sum()
plt.figure(figsize=(8, 6))
plt.pie(gender_totals, labels=gender_totals.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Previous Purchases by Gender')
plt.show()


```
